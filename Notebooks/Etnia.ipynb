```python
import pandas as pd 
```

```python
df = pd.read_csv("/Users/juanasipriano/Downloads/etnia_tahola.csv")

taholah = df[df["name"] == "Taholah, WA"].copy()

# Verificar
```

```python
taholah["Blanco"] = taholah["B03002003"]

# Negro (Black alone, Non-Hispanic)
taholah["Negro"] = taholah["B03002004"]

# Nativo (AIAN + Native Hawaiian/Pacific Islander non-hispanic)
taholah["Nativo"] = taholah["B03002005"] 

# Hispano total
taholah["Hispano"] = taholah["B03002012"]

# Otro (Asian + Other race + Two or more races)
taholah["Otro"] = (
    taholah["B03002006"] +
    taholah["B03002008"] +
    taholah["B03002009"] +
    taholah["B03002007"]

)    

```

```python
taholah["Total_Poblacion"] = taholah["B03002001"]
```

```python
taholah["%Blanco"] = taholah["Blanco"] / taholah["Total_Poblacion"] * 100
taholah["%Negro"]  = taholah["Negro"]  / taholah["Total_Poblacion"] * 100
taholah["%Nativo"] = taholah["Nativo"] / taholah["Total_Poblacion"] * 100
taholah["%Hispano"] = taholah["Hispano"] / taholah["Total_Poblacion"] * 100
taholah["%Otro"]    = taholah["Otro"]    / taholah["Total_Poblacion"] * 100

taholah
```

```python
taholah_final = taholah[[
    "geoid", "name", "Total_Poblacion",
    "Blanco", "Negro", "Nativo", "Hispano", "Otro",
    "%Blanco", "%Negro", "%Nativo", "%Hispano", "%Otro"
]]

taholah_final
```

```python
taholah_final.to_csv("taholah_race_groups_clean.csv", index=False)
```
