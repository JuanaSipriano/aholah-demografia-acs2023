```python
import pandas as pd 
```

```python
df = pd.read_csv("/Users/juanasipriano/Downloads/taholah_familia.csv")

taholah_familias = df[df["name"] == "Taholah, WA"].copy()

taholah_familias
```

```python
taholah_familias["parejas_casadas"] = taholah_familias["B11002003"]


taholah_familias["hombre_cabeza_familia"] = taholah_familias["B11002006"]


taholah_familias["mujer_cabeza_familia"] = taholah_familias["B11002009"]

taholah_familias["no_familiar"] = taholah_familias["B11002012"]

taholah_familias["total"] = taholah_familias["B11002001"]
```

```python
taholah_familias["%parejas_casadas"] = taholah_familias["parejas_casadas"] / taholah_familias["total"] * 100
taholah_familias["%hombre_cabeza_familia"] = taholah_familias["hombre_cabeza_familia"] / taholah_familias["total"] * 100
taholah_familias["%mujer_cabeza_familia"] = taholah_familias["mujer_cabeza_familia"] / taholah_familias["total"] * 100
taholah_familias["%no_familiar"] = taholah_familias["no_familiar"] / taholah_familias["total"] * 100
```
